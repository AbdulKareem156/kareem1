```python
import numpy as np
import matplotlib.pyplot as plt
a=np.array([1,2,3,4])
b=np.array([5,6,7,8])
plt.plot(a,b,c='k',marker='D')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title('Line graph')
```

```python
#plt.plot(x-axis,y-xais)
overs=[0,4,8,13,16,20]
team1=[0,30,67,98,123,145]
team2=[0,27,75,103,110,135]
plt.plot(overs,team1,label='Team1')
plt.plot(overs,team2,label='Team2')
plt.title('Analysis of cricket match')
plt.xlabel('overs')
plt.ylabel('runs')
plt.legend
```

```python
x=np.array([60,70,80,90,100])
y=np.array([250,270,260,280,300])
plt.plot(x,y)
plt.xlabel("Averge Pulse")
plt.ylabel("Calories burnt")
plt.title("Sports watch data")
plt.grid()
```

```python
#3.subplot
from random import randint
fig=plt.figure()
x1=np.random.randint(0,10,5)
x2=np.random.randint(0,10,5)
y1=np.random.randint(0,10,5)
y2=np.random.randint(0,10,5)
ax1=fig.add_subplot(339)#3r,3c,9subplot
ax2=fig.add_subplot(331)#3r,3c,1subplot
ax1.plot(x1,y1)
ax2.plot(x2,y2)
```
